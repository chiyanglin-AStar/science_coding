```python
import numpy as np
import matplotlib.pyplot as plt

N = 100000

#x = np.arange(0, 15, 0.1)
x = np.random.rand(N)
y = np.random.rand(N)
#y = np.sin(x)
#yy = np.cos(x)
#print("x:",x)
#print("y:",y)
for i in range(0,N):
  if((x[i]*x[i]+y[i]*y[i])>1):
    x[i]=0
    y[i]=0
plt.plot(x,y,'o',color='green')
#plt.plot(x, yy, color='yellow')
plt.show()
```
