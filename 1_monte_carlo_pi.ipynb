```python
import numpy as np
import matplotlib.pyplot as plt

N = 100000
pi_c = 0
x = np.random.rand(N)
y = np.random.rand(N)

#x = np.random.uniform(0,1,N)
#y = np.random.uniform(0,1,N)

for i in range(0,N):
  if((x[i]*x[i]+y[i]*y[i])>1):
    x[i]=0
    y[i]=0
  else:
    pi_c+=1
plt.plot(x,y,'o',color='green')
plt.show()
print(4*float(pi_c/N))
```
