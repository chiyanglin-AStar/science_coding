## Wavelet

A wavelet is a mathematical function used in signal processing and image compression. It is a small, localized waveform that can be used to transform and analyze signals in both time and frequency domains. The term "wavelet" is derived from "wave" and "let" (meaning small), emphasizing its ability to represent signals at various scales.

Wavelet analysis is particularly useful in applications where traditional Fourier analysis may not be ideal, such as for signals with non-stationary or time-varying characteristics. The wavelet transform breaks down a signal into its component parts at different scales, allowing for a more detailed analysis of both high and low-frequency components.

There are various types of wavelets, each with different properties and applications. Two common categories are continuous wavelet transforms (CWT) and discrete wavelet transforms (DWT). Wavelets have found applications in image and signal processing, data compression, denoising, and feature extraction, among other areas.



## Wavelet Example 1:

 Let's use the PyWavelets library in Python to perform a simple discrete wavelet transform (DWT) on a signal. If you haven't installed the PyWavelets library yet, you can do so by running:

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Generate a simple signal
signal = np.array([2, 4, 6, 8, 10, 12, 14, 16])

# Perform a single-level discrete wavelet transform (DWT)
coeffs = pywt.dwt(signal, 'db1')  # 'db1' is Daubechies wavelet with 1 vanishing moment

# Reconstruction of the signal from the wavelet coefficients
reconstructed_signal = pywt.idwt(coeffs[0], coeffs[1], 'db1')

# Plotting
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(signal, marker='o', linestyle='-', color='b')
plt.title('Original Signal')

plt.subplot(3, 1, 2)
plt.plot(coeffs[0], marker='o', linestyle='-', color='r', label='Approximation Coefficients')
plt.plot(coeffs[1], marker='o', linestyle='-', color='g', label='Detail Coefficients')
plt.legend()
plt.title('Wavelet Coefficients')

plt.subplot(3, 1, 3)
plt.plot(reconstructed_signal, marker='o', linestyle='-', color='purple')
plt.title('Reconstructed Signal')

plt.tight_layout()
plt.show()
```

In this example, we generate a simple signal and perform a single-level discrete wavelet transform using the Daubechies wavelet ('db1'). The code then plots the original signal, wavelet coefficients (approximation and detail coefficients), and the reconstructed signal.

Feel free to experiment with different signals and wavelets to see how wavelet transforms can be applied in various scenarios.
