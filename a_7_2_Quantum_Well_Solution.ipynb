## Quantum well solution

Certainly! Let’s solve the **quantum well problem**, a foundational example in quantum mechanics. We'll derive the solution for a **one-dimensional finite potential well**.

---

### **Problem Setup**

A particle of mass $m$ is confined in a one-dimensional potential well with finite depth. The potential $V(x)$ is given as:

$V(x) =
\begin{cases}
0 & \text{if } x < -L/2 \text{ or } x > L/2 \quad (\text{outside the well}) \\
-V_0 & \text{if } -L/2 \leq x \leq L/2 \quad (\text{inside the well}),
\end{cases}$

where $V_0 > 0$ is the depth of the potential well, and $L$ is the width of the well.

The time-independent Schrödinger equation is:

$-\frac{\hbar^2}{2m} \frac{d^2\psi(x)}{dx^2} + V(x)\psi(x) = E\psi(x),$

where $E$ is the energy of the particle, and we want to find the allowed energy levels and corresponding wavefunctions.

---

### **Case 1: Inside the Well** $-L/2 \leq x \leq L/2$

In this region, the potential is constant: $V(x) = -V_0$. Substituting into the Schrödinger equation gives:

$-\frac{\hbar^2}{2m} \frac{d^2\psi(x)}{dx^2} - V_0\psi(x) = E\psi(x),$

or equivalently:

$\frac{d^2\psi(x)}{dx^2} = -k^2 \psi(x),$

where $k^2 = \frac{2m}{\hbar^2} (E + V_0)$, and $E + V_0 > 0$ ensures that $k^2 > 0$.

The general solution inside the well is:

$\psi_{\text{in}}(x) = A\cos(kx) + B\sin(kx),$

where $A$ and $B$ are constants.

---

### **Case 2: Outside the Well** ($x < -L/2$ or $x > L/2$)

In this region, the potential is $V(x) = 0$. Substituting into the Schrödinger equation gives:

$-\frac{\hbar^2}{2m} \frac{d^2\psi(x)}{dx^2} = E\psi(x),$

or equivalently:

$\frac{d^2\psi(x)}{dx^2} = \kappa^2 \psi(x),$

where $\kappa^2 = \frac{2m}{\hbar^2} (V_0 - E)$, and $V_0 > E$ ensures $\kappa^2 > 0$.

The general solution outside the well is:

$\psi_{\text{out}}(x) =
\begin{cases}
C e^{\kappa x} + D e^{-\kappa x}, & x < -L/2, \\
F e^{\kappa x} + G e^{-\kappa x}, & x > L/2,
\end{cases}$

where $C, D, F, G$ are constants. However, since the wavefunction must be **normalizable**, we require $\psi(x) \to 0$ as $|x| \to \infty$, so:

- $C = 0$ for $x < -L/2$,
- $F = 0$ for $x > L/2$.

Thus, the wavefunction outside the well is:

$\psi_{\text{out}}(x) =
\begin{cases}
D e^{-\kappa x}, & x < -L/2, \\
G e^{\kappa x}, & x > L/2.
\end{cases}$

---

### **Boundary Conditions**

The wavefunction $\psi(x)$ and its derivative $\frac{d\psi(x)}{dx}$ must be **continuous** at the boundaries $x = \pm L/2$. This gives four conditions:

1. $\psi_{\text{in}}(-L/2) = \psi_{\text{out}}(-L/2)$,
2. $\frac{d\psi_{\text{in}}}{dx}(-L/2) = \frac{d\psi_{\text{out}}}{dx}(-L/2)$,
3. $\psi_{\text{in}}(L/2) = \psi_{\text{out}}(L/2)$,
4. $\frac{d\psi_{\text{in}}}{dx}(L/2) = \frac{d\psi_{\text{out}}}{dx}(L/2)$.

Substituting the solutions inside and outside the well, we get:

#### Boundary at $x = -L/2$:

$A\cos\left(-\frac{kL}{2}\right) + B\sin\left(-\frac{kL}{2}\right) = D e^{\kappa L/2},$

$-k A\sin\left(-\frac{kL}{2}\right) + k B\cos\left(-\frac{kL}{2}\right) = -\kappa D e^{\kappa L/2}.$

#### Boundary at $x = L/2$:
$A\cos\left(\frac{kL}{2}\right) + B\sin\left(\frac{kL}{2}\right) = G e^{-\kappa L/2},$

$-k A\sin\left(\frac{kL}{2}\right) + k B\cos\left(\frac{kL}{2}\right) = \kappa G e^{-\kappa L/2}.$

---

### **Quantization Condition**

For non-trivial solutions (bound states), the determinant of the system must vanish. After simplifying the algebra, this results in a **transcendental equation** for the allowed energies $E$:

#### Even Solutions ($\psi_{\text{in}}(x) \propto \cos(kx)$):

$k \tan\left(\frac{kL}{2}\right) = \kappa.$

#### Odd Solutions ($\psi_{\text{in}}(x) \propto \sin(kx)$):

$k \cot\left(\frac{kL}{2}\right) = -\kappa.$

Here:
$k = \sqrt{\frac{2m}{\hbar^2}(E + V_0)}, \quad \kappa = \sqrt{\frac{2m}{\hbar^2}(V_0 - E)}.$

---

### **Key Observations**

1. The transcendental equations must be solved numerically for $E$.
2. The depth $V_0$ and width $L$ determine the number of bound states.
3. The wavefunctions alternate between **even** and **odd parity** as you go up in energy levels.

---

### **Numerical Solution**

The transcendental equations can be solved numerically using Python's `scipy.optimize.root_scalar` to find $E \). Let me know if you’d like the numerical implementation!

Here is the Python code used to generate the plot of the finite potential well solutions:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar

# Constants
hbar = 1.0545718e-34  # Reduced Planck's constant (J.s)
m = 9.10938356e-31    # Mass of electron (kg)
L = 1e-9              # Width of the well (m)
V0 = 50 * 1.60218e-19 # Depth of the well (50 eV in Joules)

# Define the transcendental equations for even and odd solutions
def even_eq(E):
    k = np.sqrt(2 * m * (E + V0)) / hbar
    kappa = np.sqrt(2 * m * (V0 - E)) / hbar
    return k * np.tan(k * L / 2) - kappa

def odd_eq(E):
    k = np.sqrt(2 * m * (E + V0)) / hbar
    kappa = np.sqrt(2 * m * (V0 - E)) / hbar
    return k * np.tan(k * L / 2) + kappa

# Function to find valid brackets for root finding
def find_brackets(func, x_range, step_size):
    brackets = []
    x_vals = np.arange(x_range[0], x_range[1], step_size)
    for i in range(len(x_vals) - 1):
        if func(x_vals[i]) * func(x_vals[i + 1]) < 0:  # Sign change indicates a root
            brackets.append((x_vals[i], x_vals[i + 1]))
    return brackets

# Scan for valid brackets
brackets_even = find_brackets(even_eq, [-V0, 0], 1e-21)
brackets_odd = find_brackets(odd_eq, [-V0, 0], 1e-21)

# Solve for energy levels using the identified brackets
energy_levels = []
for eq, brackets in zip([even_eq, odd_eq], [brackets_even, brackets_odd]):
    for bracket in brackets:
        sol = root_scalar(eq, bracket=bracket, method='brentq')
        if sol.converged and sol.root not in energy_levels:
            energy_levels.append(sol.root)

energy_levels = sorted(energy_levels)

# Initialize x and potential profile
x = np.linspace(-2 * L, 2 * L, 1000)
V = np.piecewise(x, [x < -L / 2, (x >= -L / 2) & (x <= L / 2), x > L / 2],
                 [0, -V0, 0])

# Compute wavefunctions
wavefunctions = []
for E in energy_levels:
    k = np.sqrt(2 * m * (E + V0)) / hbar
    kappa = np.sqrt(2 * m * (V0 - E)) / hbar
    psi = np.zeros_like(x)

    # Inside the well
    inside = (x >= -L / 2) & (x <= L / 2)
    if np.isclose(k * np.tan(k * L / 2), kappa):  # Even solution
        psi[inside] = np.cos(k * (x[inside] + L / 2))
    else:  # Odd solution
        psi[inside] = np.sin(k * (x[inside] + L / 2))

    # Outside the well
    outside_left = x < -L / 2
    outside_right = x > L / 2
    psi[outside_left] = np.exp(kappa * (x[outside_left] + L / 2))
    psi[outside_right] = np.exp(-kappa * (x[outside_right] - L / 2))

    # Normalize wavefunction
    psi /= np.sqrt(np.trapz(psi**2, x))
    wavefunctions.append(psi)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x * 1e9, V / 1.60218e-19, label="Potential (eV)", color="black", linewidth=2)
for i, psi in enumerate(wavefunctions):
    plt.plot(x * 1e9, psi + energy_levels[i] / 1.60218e-19, label=f"n={i+1}")

# Labels and legend
plt.xlabel("Position (nm)")
plt.ylabel("Energy (eV)")
plt.title("Finite Potential Well Solutions")
plt.legend()
plt.grid()
plt.show()
```

### Key Sections:
1. **Potential Well Definition**: The potential well $V(x)$ is defined piecewise.
2. **Energy Levels**: Energy levels are found using the transcendental equations for even and odd solutions.
3. **Wavefunctions**: Wavefunctions are calculated and normalized for each energy level.
4. **Plot**: The potential well and wavefunctions are plotted together, with energy levels indicated.
