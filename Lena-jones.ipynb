```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,0.1,1000]) # xpoint = r
ypoints = 1/ pow(xpoints,12) - 1 / pow(xpoints,6) # ypoint = 1/r^12 - 1/r^6

plt.plot(xpoints, ypoints)
plt.show()
```
