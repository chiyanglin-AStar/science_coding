```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
num_particles = 1000  # Number of particles in the beam
spin_up_prob = 0.5  # Probability of spin-up (equal distribution assumed)

# Generate random initial positions for particles in 2D (x, z plane)
x_positions = np.random.uniform(-1, 1, num_particles)  # x positions (beam width)
z_positions = np.zeros(num_particles)  # All start at z = 0

# Generate spin states for each particle
# +1 for spin-up, -1 for spin-down
spin_states = np.random.choice([1, -1], size=num_particles, p=[spin_up_prob, 1 - spin_up_prob])

# Magnetic field gradient (affects z-direction displacement)
field_gradient = 1.0  # Arbitrary units
z_displacement = spin_states * field_gradient  # Spin-dependent displacement in z

# Final positions after passing through the Stern-Gerlach apparatus
z_final_positions = z_positions + z_displacement

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(x_positions[spin_states == 1], z_final_positions[spin_states == 1],
            color='blue', label='Spin Up (+/2)', alpha=0.6)
plt.scatter(x_positions[spin_states == -1], z_final_positions[spin_states == -1],
            color='red', label='Spin Down (-/2)', alpha=0.6)

# Formatting the plot
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.7)
plt.xlabel('X Position (Beam Width)')
plt.ylabel('Z Position (Deflection)')
plt.title('Stern-Gerlach Experiment Simulation')
plt.legend()
plt.grid(True)
plt.show()
```
